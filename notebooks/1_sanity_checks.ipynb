### Making sanity checks for input data

This notebooks facilitates the implementation of sanitu checks for the input raw data, including:

- Check Nulls
- Check columns
- Check descriptors
- Check counts

```python
import pandas as pd
```

- Checking cell viability file

```python
df_cell_viability = pd.read_excel("../raw_data/Cell Viability.xlsx")
df_cell_viability.shape
```

```python
df_cell_viability.describe()
```

```python
df_cell_viability.head()
```

```python
df_cell_viability["Cell line"].value_counts()
```

```python
df_cell_viability["Enzyme"].value_counts()
```

```python
df_cell_viability["Treatment (%)"].value_counts()
```

```python
grid = (
    df_cell_viability
    .groupby(["Cell line", "Treatment (%)"])["Cell viability (% of control)"]
    .mean()
    .reset_index()
)
grid
```
